# Exploratory Causal Analysis

This notebook demonstrates how to use the Causal Engine for structure learning and inference.

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from src.causal_graph.structure_learning import learn_graph_from_data

%matplotlib inline
```

```python
# Generate synthetic data
np.random.seed(42)
n_samples = 1000
X = np.random.randn(n_samples)
Y = 3 * X + np.random.randn(n_samples) * 0.5
Z = 2 * Y - X + np.random.randn(n_samples) * 0.5

df = pd.DataFrame({'X': X, 'Y': Y, 'Z': Z})
```

```python
# Learn Graph
G = learn_graph_from_data(df)
nx.draw(G, with_labels=True)
```
